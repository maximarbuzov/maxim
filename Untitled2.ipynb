```python
from sklearn.preprocessing import normalize
```

Скопируем часть датафрейма с числовыми значениями в отдельный датафрейм для дальнейшего анализа.

```python
df=data[['agricultural_land', 'forest_area', 'land_area', 'rural_land', 'urban_land']].copy()
```

```python
df
```

```python
df.info()
```

Видим пропуск данных в столбце agricultural_land, локализуем его для принятия дальнейшего решения

```python
df.loc[df.agricultural_land.isna()]
```

Т.к. данные в остальных столбцах могут быть существенны, пропуск не удаляем, а заменяем значение близким к нулю

```python
df.loc[df.agricultural_land.isna()]=0.01
```

```python
df.info()
```

```python
df.describe()
```

Видим наличие нулевых значений в столбцах forest_area и urban_land. Так же локализуем и заменяем на значения близкие к нулю для дальнейшей обработки.

```python
df.loc[df.urban_land==0]
```

```python
df.loc[df.urban_land==0]=0.01
```

```python
df.loc[df.forest_area==0]
```

```python
df.loc[df.forest_area==0]=0.01
```

```python
df.describe()
```

Нормализуем данные

```python
normalized_data1 = normalize(df, norm='l1', axis=1)
normalized_df1 = pd.DataFrame(normalized_data1, columns=df.columns)
```

```python
normalized_df1
```

```python
normalized_data2 = normalize(df, norm='l2', axis=1)
normalized_df2 = pd.DataFrame(normalized_data2, columns=df.columns)
```

```python
normalized_df2
```

```python
normalized_data3 = normalize(df, norm='max', axis=1)
normalized_df3 = pd.DataFrame(normalized_data3, columns=df.columns)
```
