```python
data2=pd.read_csv("http://intersyn.net/Life_Table1.csv")
```

```python
data2.head()
```

```python
data2.info()
```

```python
data2.Location.unique()
```

```python
data2=data2[data2.Location=='United States of America']
```

```python
data2
```

```python
data2.AgeGrp.unique()
```

```python
data_1 = data2.groupby('Time')['ex'].mean()
data_2 = data2.groupby('Time')['mx'].mean()

fig, ax1 = plt.subplots()

ax1.set_xlabel('Time')
ax1.set_ylabel('ex', color = 'orange')
ax1.plot(data_1, color = 'orange')
ax1.tick_params(axis ='y', labelcolor = 'orange')

ax2 = ax1.twinx()

ax2.set_ylabel('mx', color = 'forestgreen')
ax2.plot(data_2, color = 'forestgreen')
ax2.tick_params(axis ='y', labelcolor = 'forestgreen')

plt.show()
```
