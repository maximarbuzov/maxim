```python
from scipy.stats.contingency import chi2_contingency
```

```python
#Ho - статистически значимая связь между полом и уровнем дохода отсутствует
contingency_table=pd.crosstab(data['sex'],data['class'])
```

```python
contingency_table.head()
```

```python
chi2_contingency(contingency_table)
```

```python
# p-value < 0.05, следовательно связь между полом и уровнем дохода существует
```
