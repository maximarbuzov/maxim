```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
data = pd.read_csv("http://intersyn.net/Churn.csv")
```

```python
data
```

```python
data=data.dropna()
```

```python
data
```

```python
card_clients=data.Exited[data.HasCrCard==1]
no_card_clients=data.Exited[data.HasCrCard==0]
```

```python
card_clients
```

```python
no_card_clients
```

```python
len(card_clients)
```

```python
len(no_card_clients)
```

```python
conversion_rate_A=np.count_nonzero(no_card_clients)/len(no_card_clients)
conversion_rate_B=np.count_nonzero(card_clients)/len(card_clients)
```

```python
conversion_rate_A
```

```python
conversion_rate_B
```

```python
from scipy.stats import beta
```

```python
posterior_A=beta(conversion_rate_A*len(no_card_clients)+1,(1-conversion_rate_A)*len(no_card_clients)+1)
```

```python
posterior_B=beta(conversion_rate_B*len(card_clients)+1,(1-conversion_rate_B)*len(card_clients)+1)
```

```python
posterior_difference=posterior_B.rvs(10000)-posterior_A.rvs(10000)
```

```python
plt.plot(np.linspace(0,1,900),posterior_A.pdf(np.linspace(0,1,900)))
plt.plot(np.linspace(0,1,900),posterior_B.pdf(np.linspace(0,1,900)))
plt.xlim(0.15,0.25)
```

```python
sns.kdeplot(posterior_difference)
```
