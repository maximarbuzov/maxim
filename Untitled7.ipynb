```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
#Задание 1.
#Загрузка и первичный анализ датафрейма
data=pd.read_excel('Nasdaq.xlsx')
```

```python
data.info()
```

```python
data.columns=data.columns.str.replace(' ','_')
```

```python
data.describe()
```

```python
#Задание 2.
#Выбор случайной компании
np.random.seed(0)
company_number=np.random.randint(1, len(data))
company_number
```

```python
#Вывод данных о компании
data[['Symbol','IPO_Year','Country','Sector','Industry']].loc[company_number]
```

```python
data.corr()
```

```python
#Задание 3
#Оценить степень связи между двумя категориальными переменными: страной и индустрией
#Используем критерий Пирсона
from scipy.stats.contingency import chi2_contingency
```

```python
#Ho - статистически значимая связь между страной и индустрией отсутствует
contingency_table=pd.crosstab(data['Country'],data['Industry'])
```

```python
contingency_table
```

```python
chi2_contingency(contingency_table)
```

```python
# p-value < 0.05, следовательно связь между страной и индустрией существует
```

```python
#Задание 4.
#Рассмотреть, как распределены доли рыночной стоимости компаний из этого сектора. Сравнить с распределениями всех остальных секторов.
```

```python
sector=data.Sector.loc[company_number]
sector
```

```python
sector_companies=data[data.Sector==sector]
sector_companies.head()
```

```python
sns.histplot(data=sector_companies.Market_Cap,bins=50,kde=True,color='skyblue')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('Miscellaneous - Market_Cap')
```
