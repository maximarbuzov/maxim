```python
stock_prices=yf.download(tickers='OLPX', period='1500d',interval='1d',group_by='column')
```

```python
stock_prices
```

```python
stock_prices['Adj Close'].plot()
```

```python
plt.figure(figsize=(36,20))

# "up" dataframe will store the stock_prices
# when the closing stock price is greater
# than or equal to the opening stock prices
up = stock_prices[stock_prices.Close >= stock_prices.Open]

# "down" dataframe will store the stock_prices
# when the closing stock price is
# lesser than the opening stock prices
down = stock_prices[stock_prices.Close < stock_prices.Open]

# When the stock prices have decreased, then it
# will be represented by red color candlestick
col1 = 'green'

# When the stock prices have increased, then it
# will be represented by green color candlestick
col2 = 'red'

# Setting width of candlestick elements
width = .6
width2 = .06

# Plotting up prices of the stock
plt.bar(up.index, up.Close-up.Open, width, bottom=up.Open, color=col1)
plt.bar(up.index, up.High-up.Close, width2, bottom=up.Close, color=col1)
plt.bar(up.index, up.Low-up.Open, width2, bottom=up.Open, color=col1)

# Plotting down prices of the stock
plt.bar(down.index, down.Close-down.Open, width, bottom=down.Open, color=col2)
plt.bar(down.index, down.High-down.Open, width2, bottom=down.Open, color=col2)
plt.bar(down.index, down.Low-down.Close, width2, bottom=down.Close, color=col2)

# rotating the x-axis tick labels at 30degree
# towards right
plt.xticks(rotation=30, ha='right')

# displaying candlestick chart of stock data
# of a week
plt.show()
```

Видим, что с начала обращения акций на бирже до настоящего времени, их цена снизилась с 25 до 1,5 $

```python
#Задание 6.
#Рассчитать индекс индустрии за последние 20 лет (можно меньше) на основе 10 крупнейших компаний.
#Сравнить динамику индекса и курса акций выбранной компании.
```

```python
industry_companies=sector_companies[sector_companies.Industry=='Diversified Commercial Services']
industry_companies.head()
```

```python
industry_companies=industry_companies.sort_values(by="Market_Cap", ascending=False)
industry_companies.head()
```

```python
symbols=industry_companies.Symbol[:10]
symbols
```

```python
prices=pd.DataFrame(columns=symbols)
prices
```

```python
for i in symbols:
  prices[i]=yf.download(tickers=i, period='3650d',interval='1d',group_by='column')['Adj Close']
prices
```

```python
prices.describe()
```

```python
prices.info()
```

```python
prices=prices.dropna(axis=1)
```

```python
prices
```

```python
#Расчет индекса
#Расчет средних цен на акции за период:
period_avg_prices = []
for day in range(528):
  daily_prices = [prices[company][day] for company in prices]
  avg_price = sum(daily_prices) / len(daily_prices)
  period_avg_prices.append(avg_price)
#Расчет индекса (базовое значение индекса 100)
base_index = 100
period_index = [base_index]
for i in range(1, len(period_avg_prices)):
  index_change = (period_avg_prices[i] - period_avg_prices[i - 1]) / period_avg_prices[i - 1] * 100
  period_index.append(period_index[-1] + index_change)
# Вывод результатов
for day, index_value in enumerate(period_index):
    print(f"День {day + 1}: Индекс = {index_value:.2f}")
```

```python
prices['Period_Index']=period_index
```

```python
prices
```

```python
prices.info()
```

```python
prices['OLPX'].plot()
prices['CASY'].plot()
prices['Period_Index'].plot()
plt.legend()
plt.show()
```

```python
data_1 = prices['OLPX']
data_2 = prices['Period_Index']

fig, ax1 = plt.subplots()

ax1.set_xlabel('Date')
ax1.set_ylabel('OLPX', color = 'deepskyblue')
ax1.plot(data_1, color = 'deepskyblue')
ax1.tick_params(axis ='y', labelcolor = 'deepskyblue')

ax2 = ax1.twinx()

ax2.set_ylabel('Period_Index', color = 'forestgreen')
ax2.plot(data_2, color = 'forestgreen')
ax2.tick_params(axis ='y', labelcolor = 'forestgreen')

plt.show()
```

```python
data_1 = prices['CASY']
data_2 = prices['Period_Index']

fig, ax1 = plt.subplots()

ax1.set_xlabel('Date')
ax1.set_ylabel('OLPX', color = 'orange')
ax1.plot(data_1, color = 'orange')
ax1.tick_params(axis ='y', labelcolor = 'orange')

ax2 = ax1.twinx()

ax2.set_ylabel('Period_Index', color = 'forestgreen')
ax2.plot(data_2, color = 'forestgreen')
ax2.tick_params(axis ='y', labelcolor = 'forestgreen')

plt.show()
```
