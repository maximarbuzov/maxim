```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('All Countries.csv')
```

```python
data
```

Скопируем часть датафрейма с числовыми значениями в отдельный датафрейм для дальнейшего анализа.

```python
df=data[['country','agricultural_land', 'forest_area', 'land_area', 'rural_land', 'urban_land']].copy()
```

```python
df
```

```python
df.loc[df.agricultural_land.isna()]=0.01
df.loc[df.urban_land==0]=0.01
df.loc[df.forest_area==0]=0.01
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
sns.lmplot(x='land_area',y='forest_area',data=df)
```

```python
sns.histplot(data=df.agricultural_land,bins=500,kde=True,color='skyblue')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('agricultural_land')
```

```python
sns.histplot(data=df.forest_area,bins=500,kde=True,color='skyblue')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('forest_area')
```

```python
sns.histplot(data=df.land_area,bins=500,kde=True,color='skyblue')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('land_area')
```

```python
sns.histplot(data=df.rural_land,bins=500,kde=True,color='skyblue')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('rural_land')
```

```python
sns.histplot(data=df.urban_land,bins=500,kde=True,color='skyblue')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('urban_land')
```
